```python


def fit_predict_eval(model, features_train, features_test, target_train, target_test):
    if isinstance(model, KNeighborsClassifier):
        knn_model = model
        knn_model.fit(features_train, target_train)
        accuracy_score = knn_model.score(features_test, target_test)
        print(f'Model: {model}\nAccuracy: {round(accuracy_score, 3)}\n')
        return round(accuracy_score, 3)

    if isinstance(model, RandomForestClassifier):
        rf_model = model
        rf_model.fit(features_train, target_train)
        accuracy_score = rf_model.score(features_test, target_test)
        print(f'Model: {model}\nAccuracy: {round(accuracy_score, 3)}\n')
        return round(accuracy_score, 3)


(x_train, y_train), _ = keras.datasets.mnist.load_data(path="mnist.npz")
x_train_two = np.reshape(x_train, (x_train.shape[0], x_train.shape[1] * x_train.shape[2]))

data_subset = x_train_two[:6000]
target_subset = y_train[:6000]
X_train, X_test, Y_train, Y_test = train_test_split(data_subset, target_subset, test_size=0.3, random_state=40)

normalizer = Normalizer()
X_train_normalized = normalizer.transform(X_train)
X_test_normalized = normalizer.transform(X_test)

param_grid_knn = {
    'n_neighbors': [3, 4,5,6],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'brute']
}

param_grid_rf = {
    "n_estimators": [500, 600],
    "max_features": ["sqrt", "log2"],
    "class_weight": ["balanced", "balanced_subsample"],
}

knn_classifier = KNeighborsClassifier()
grid_search_knn = GridSearchCV(estimator=knn_classifier,
                               param_grid=param_grid_knn,
                               scoring='accuracy',
                               n_jobs=-1)

grid_search_knn.fit(X_train_normalized, Y_train)
best_params_knn = grid_search_knn.best_estimator_
accuracy_knn = grid_search_knn.best_score_

rf_classifier = RandomForestClassifier(random_state=40)
grid_search_rf = GridSearchCV(estimator=rf_classifier,
                              param_grid=param_grid_rf,
                              scoring='accuracy',
                              n_jobs=-1)

grid_search_rf.fit(X_train, Y_train)
best_params_rf = grid_search_rf.best_estimator_
accuracy_rf = grid_search_rf.best_score_

print("K-nearest neighbours algorithm")
print(f"best estimator: {best_params_knn}")
print(f"accuracy: {accuracy_knn}")
print("")
print("Random forest algorithm")
print(f"best estimator: {best_params_rf}")
print(f"accuracy: {accuracy_rf}")
```
